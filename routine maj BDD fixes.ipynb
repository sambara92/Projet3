ROUTINE MAJ BDD AERONEFS
sur BDD 'log_vol' si etat_voyant != 0:
    alors sur AERONEFS en_maintenance == oui
    alors sur AERONEFS last_maintenance == jour_vol
    si etat_voyant == 1:
        alors sur AERONEFS end_maint = last_maint + 1
    si etat_voyant == 2:
        alors sur AERONEFS end_maint = last_maint + 7
    si etat_voyant == 3:
    alors sur AERONEFS end_maint = last_maint + 14

```python
from datetime import datetime, timedelta
log_vols_06_05['jour_vol'] = pd.to_datetime(log_vols_06_05['jour_vol'])

#boucle pour maj en maintenance et début 
for index,row in aeronefs.iterrows():
    if row['ref_aero'] in (dico_voyant[1]):
        aeronefs.at[index, 'en_maintenance'] = True
        aeronefs.at[index, 'last_maint'] = jour_vol
        aeronefs.at[index, 'end_maint'] = jour_vol + timedelta(days=1)
        aeronefs['cout'] = 1 * 15000
    if row['ref_aero'] in (dico_voyant[2]):
        aeronefs.at[index, 'en_maintenance'] = True
        aeronefs.at[index, 'last_maint'] = jour_vol
        aeronefs.at[index, 'end_maint'] = jour_vol + timedelta(days=7)
        aeronefs['cout'] = 7 * 15000
    if row['ref_aero'] in (dico_voyant[3]):
        aeronefs.at[index, 'en_maintenance'] = True
        aeronefs.at[index, 'last_maint'] = jour_vol
        aeronefs.at[index, 'end_maint'] = jour_vol + timedelta(days=14)
        aeronefs['cout'] = 14 * 15000       
```
