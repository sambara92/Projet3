REFLEXION SUR STRUCTURE DE LA BDD HISTORIQUES DES MESURES JOURNALIERES

```python
import pandas as pd
```

```python
degra_06_02 = pd.read_csv(r"C:\Users\MAITE\Wild Code School\SkyAnalytics\BDD\degradations_2024-06-02.csv")
degra_06_03 = pd.read_csv(r"C:\Users\MAITE\Wild Code School\SkyAnalytics\BDD\degradations_2024-06-03.csv")
degra_06_04 = pd.read_csv(r"C:\Users\MAITE\Wild Code School\SkyAnalytics\BDD\degradations_2024-06-04.csv")
degra_06_05 = pd.read_csv(r"C:\Users\MAITE\Wild Code School\SkyAnalytics\BDD\degradations_2024-06-05.csv")
```

```python
#stocker les nouvelles mesures du jour
day_06_03 = degra_06_03.loc[degra_06_03['measure_day'] == '2024-06-03', :]
day_06_04 = degra_06_04.loc[degra_06_04['measure_day'] == '2024-06-04', :]
day_06_05 = degra_06_05.loc[degra_06_05['measure_day'] == '2024-06-05', :]
```

```python
#creer 2 colonnes vides sur le 1er fichier pour stocker les maj
degra_06_02['usure_next'] = 0
degra_06_02['next_measure_day'] = 0
```

```python
#faire un dico avec pour paire clé/valeur 'compo_concerned' / 'usure' pour chaques mesures du jour
dico_06_03 = dict(zip(day_06_03['compo_concerned'], day_06_03['usure_nouvelle']))
dico_06_04 = dict(zip(day_06_04['compo_concerned'], day_06_04['usure_nouvelle']))
dico_06_05 = dict(zip(day_06_05['compo_concerned'], day_06_05['usure_nouvelle']))
```

```python
#itérer dans chaque dico pour mettre à jour le fichier central = degra_06-02
#si clé dico = valeur colonne ref_composant alors nouvelle colonne usure maj avec valeur dico
#si clé dico = valeur colonne ref_composant alors nouvelle colonne next_measure_day maj avec valeur du jour
for key, value in dico_06_03.items():
    degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next'] = value
    degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_measure_day'] = '2024-06-03'
```

```python
degra_06_02.sample(10)
```

```python
#variante avec vérification si colonne pas déjà remplie
for key, value in dico_06_04.items():
    if degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next'].values[0] != 0:
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next'] = value
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_measure_day'] = '2024-06-04'
    else:
        degra_06_02['usure_next_2'] = 0
        degra_06_02['next_2_measure_day'] = 0
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next_2'] = value
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_2_measure_day'] = '2024-06-04'
```

```python
degra_06_02
```

```python
#variante avec vérification si 2 précédentes colonnes pas déjà remplies
for key, value in dico_06_05.items():
    if degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next'].values[0] != 0:
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next'] = value
        degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_measure_day'] = '2024-06-05'
    else:
        if degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next_2'].values[0] != 0:
            degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next_2'] = value
            degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_2_measure_day'] = '2024-06-05'
        else:
            degra_06_02['usure_next_3'] = 0
            degra_06_02['next_3_measure_day'] = 0
            degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'usure_next_3'] = value
            degra_06_02.loc[degra_06_02['compo_concerned'] == str(key), 'next_3_measure_day'] = '2024-06-05'
```

```python
#creer une fonction pour automatiser le processus
def nettoyage_vols(bdd):
    #diviser le dico sensor_data en autant de colonnes que de variables
    #jointure avec le data set principal
    bdd = bdd.join(bdd['sensor_data'].str.split(",", expand=True))
    #renommer les colonnes
    bdd = bdd.rename(columns={0:'temperature'})
    bdd = bdd.rename(columns={1:'pressure'})
    bdd = bdd.rename(columns={2:'vibrations'})

    #nettoyage des nouvelles colonnes splitées
    bdd['temperature'] = bdd['temperature'].str.replace('{', '').str.replace('temp', '').str.replace('°C', '').str.replace(':', '').str.replace('\'', '')
    bdd['pressure'] = bdd['pressure'].str.replace('pressure', '').str.replace('hPa', '').str.replace(':', '').str.replace('\'', '')
    bdd['vibrations'] = bdd['vibrations'].str.replace('vibrations', '').str.replace('m/s²', '').str.replace(':', '').str.replace('\'', '').str.replace('}', '')

    #suppression colonne inutiles
    bdd = bdd.drop(columns='sensor_data')
    return bdd
```

```python
vols_06_02 = nettoyage_vols(vols_06_02)
```

```python
vols_06_03 = nettoyage_vols(vols_06_03)
```

```python
vols_06_04 = nettoyage_vols(vols_06_04)
```

```python
vols_06_05 = nettoyage_vols(vols_06_05)
```

```python
#pour conserver un historique de tous les vols il faut ensuite concatener les bdd ensemble
```

```python
histo_vols = pd.concat([vols_06_02, vols_06_03, vols_06_04, vols_06_05])
```
